```python
#Scraper for Gorden Ramsy
from urllib.request import urlopen
from bs4 import BeautifulSoup

#url_to_scrape = ("https://parade.com/616643/lindsaylowe/10-of-gordon-ramsays-funniest-quotes-and-insults-this-isnt-pizza-this-is-a-mistake/")
url_to_read = "https://parade.com/616643/lindsaylowe/10-of-gordon-ramsays-funniest-quotes-and-insults-this-isnt-pizza-this-is-a-mistake/"

request_page = urlopen(url_to_read)
page_html = request_page.read()
request_page.close()

html_soup = BeautifulSoup(page_html, "html.parser")

# print(html_soup.body)

Gorden_insults = html_soup.find_all("li")

print(Gorden_insults)
#I think the problem is that we have not defined what goes inside of []
lis = []
for sent in Gorden_insults:
    lis.append(sent.text)

print(lis)

# soup = bs(page.content, 'html.parser')
# print(soup.prettify())
```
